```python
from IPython.display import display, HTML
def show_html(src):
    return display(HTML(src), metadata=dict(isolated=True))
```

```python
html_tmpl = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="utf-8">
    <title>{title}</title>
    <style>
    html {{ min-height: 32em; overflow: hidden; }}
    </style>
</head>
<body>
{content}
</body>
</html>"""
```

```python
#get text + length
myText = open('spiraltext.txt').read()
# print(myText)
strLen = len(myText)
```

```python
#split text into substrings of increasing length

subLen = 3 #starting substring length, increases by 2 every 2 substrings
subStrs = [] #an empty list to hold substrings
letCounter = 0 #a counter to store the current letter #

while letCounter < strLen: #for every letter in original text:
    for i in range(2):
        subStr = myText[letCounter:letCounter+subLen]
        # print(subStr)
        subStrs.append(subStr)
        letCounter+=subLen
    subLen+=2
print(subStrs)
```

```python
def mkdiv(content, **kwargs):
    if 'position' not in kwargs:
        kwargs['position'] = 'absolute'
    style_str = ' '.join([": ".join((k.replace('_', '-'), v))+";" for k, v in kwargs.items()])
    return f"<div style='{style_str}'>{content}</div>"
```

```python

divs = []
y = 50
x = 50
rot = 0
subCounter = 0


for currentSub in range(len(subStrs)): 
# for currentSub in range(6): 
    sub = subStrs[subCounter] #get variable for current sub
    offset = len(sub)*.8
    this_div = mkdiv(sub,
                     top=f"{y}%",
                     left=f"{x}%",
                     font_size="12pt",
                     font_family='courier new',
                     transform=f"rotate({rot}deg)")
    divs.append(this_div)

    rot-=90
    if subCounter%4 == 0:
        x+=offset
    elif subCounter%4 == 1:
        y-=len(sub)*1.5
    elif subCounter%4 == 2:
        x-=offset
    else:
        y+=len(sub)*1.5
    
    subCounter+=1
html_src = html_tmpl.format(title="Stars", content="".join(divs))
show_html(html_src)
```
